```python
def bar_plot(variable):
    var=sa[variable]
    varValue=var.value_counts()
    mp.figure(figsize=(9,3))
    mp.bar(varValue.index,varValue)
    mp.xticks(varValue.index,varValue.index.values)
    mp.ylabel("Frequency")
    mp.title(variable)
    mp.show()
    print("{}:\n{}".format(variable,varValue))
```

```python
columns_plot=['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm','Species']
for i in columns_plot:
    bar_plot(i)
```
